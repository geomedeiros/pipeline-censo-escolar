## Extraindo os Datasets
Como primeira etapa, será extraindo da fonte e carregado todos os dados necessários no ambiente do Databricks.

```python
# Obtém o ano atual -1 para referência na coleta dos dados
year_ref = datetime.now().year - 1

# Caminho do catálogo onde os arquivos CSV serão armazenados
volume_path = "/Volumes/workspace/default/censo_escolar/"

# Lista das tabelas a serem carregadas no ambiente
target_tables = [
    f"Tabela_Turma_{year_ref}.csv",
    f"Tabela_Escola_{year_ref}.csv"
]
```

```python
%sql
-- Cria o volume de destino dos arquivos CSV caso ainda não exista
CREATE VOLUME IF NOT EXISTS workspace.default.censo_escolar
```

```python
# Obtém os arquivos que já estão no catálogo
volume_files = os.listdir(volume_path)

# Verifica se as tabelas alvo já estão presentes no catálogo
result = all(item in volume_files for item in target_tables)
print(result)
```

Upload e carregamento do arquivo ZIP do governo (Caso os dados não estejam carregados no Catálogo)

```python
if not result:

    print("INICIANDO PROCESSO DE DOWNLOAD DA FONTE DE DADOS...\n")

    # ================================================
    # Primeira Etapa: Dowload do arquivo .ZIP (Via http)
    # ================================================

    data_url = f"https://download.inep.gov.br/dados_abertos/microdados_censo_escolar_{year_ref}_.zip" # URL de dowload no site do governo
    zip_path = f"/tmp/microdados_censo_escolar_{year_ref}_.zip" # Caminho de destino para o zip à ser baixado

    # Faz a requisição HTTP para baixar o arquivo
    with requests.get(data_url, stream=True) as r:
        r.raise_for_status()
        with open(zip_path, "wb") as f:
            for chunk in r.iter_content(8192):
                f.write(chunk)

    print("1º Etapa: Arquivo ZIP baixado")

    # ================================================
    # Segunda Etapa: Extração do arquivo .ZIP
    # ================================================

    extract_path = "/tmp/extracted" # Caminho de destino para os dados extraídos do zip

    # Extrai o arquivos ZIP
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

    print("2º Etapa: Arquivos extraídos:", os.listdir(extract_path))

    # ================================================
    # Terceira Etapa: Listagem dos arquivos .CSV na pasta de dados
    # ================================================

    tables_path = extract_path + f"/microdados_censo_escolar_{year_ref}/dados/" # Caminho onde estão os arquivos .CSV conforme padrão gov

    # Lista os arquivos CSV encontrados
    print("3º Etapa: CSVs encontrados: \n")
    for file in os.listdir(tables_path):
        if file.endswith(".csv"):
            csv_file = os.path.join(tables_path, file)
            print(csv_file)
    print("\n")

    # ================================================
    # Quarta Etapa: Carregamento dos .CSV desejados ao Volume
    # ================================================

    import shutil
        
    print("4º Etapa: ")
    for table in target_tables:
        source_path = tables_path + table # Caminho de origem
        destination_path = volume_path + table # Caminho de destino
        
        # Remove se já existir
        if os.path.exists(destination_path):
                os.remove(destination_path)

        # Copia o CSV do /temp para o catálogo
        shutil.copy(source_path, destination_path)

        print(
        f"""
Arquivo CSV adicionado ao catálogo com sucesso!!
Arquivo: {source_path}
Caminho: {destination_path}"""
        )
    print("\n")
    
    # ================================================
    # Última Etapa: Limpeza da pasta /temp
    # ================================================

    # Remove os arquivos desnecessários da pasta temporária
    try:
        os.remove(zip_path)
        shutil.rmtree(extract_path)
        print("5º Etapa: Arquivos temporários removidos com sucesso!")
    except:
        print("5º Etapa: Os arquivos já não existem na pasta temporária")

    # ================================================

    print("\nFONTE DE DADOS BAIXADA COM SUCESSO!!!")
```

## Leitura dos Datasets
Realiza a leitura dos arquivos CSV no Volume e transformam em spark dataframes para que sejam manipulados no pipeline.

```python
df_raw_class = spark.read.csv(
    volume_path + f"Tabela_Turma_{year_ref}.csv", 
    header=True, 
    inferSchema=True, 
    sep=';', 
    encoding='latin1'
)
df_raw_school = spark.read.csv(
    volume_path + f"Tabela_Escola_{year_ref}.csv", 
    header=True, 
    inferSchema=True, 
    sep=';', 
    encoding='latin1'
    )

print(f"✅ Datasets carregados com sucesso!")
print(f"📊 Registros Tabela Escola: {df_raw_school.count():,}")
print(f"📊 Registros Tabela Turma: {df_raw_class.count():,}")
```

## Camada BRONZE
Salva os dados brutos com metadados de controle.

```python
from pyspark.sql import functions as f

# Adiciona 3 colunas de controle
df_bronze_school = df_raw_school \
    .withColumn("_ingestao_timestamp", f.current_timestamp()) \
    .withColumn("_fonte", f.lit(f"gov/Tabela_Escola{year_ref}.csv")) \
    .withColumn("_pipeline_version", f.lit("1.0.0"))

df_bronze_class = df_raw_class \
    .withColumn("_ingestao_timestamp", f.current_timestamp()) \
    .withColumn("_fonte", f.lit(f"gov/Tabela_Turma{year_ref}.csv")) \
    .withColumn("_pipeline_version", f.lit("1.0.0"))
```

```python
# Caminho para as tabelas gerenciadas na camada BRONZE
bronze_path_school = "workspace.default.bronze_escola" 
bronze_path_class = "workspace.default.bronze_turma"

# Define o modo de escrita
def get_write_mode_by_year(target_path, year_ref):
    year_on_table = spark.table(target_path).select("NU_ANO_CENSO").filter(f.col("NU_ANO_CENSO") == year_ref).limit(1)
    if year_on_table.count() == 1:
        return "overwrite"
    else:
        return "append"

# Salva como tabela gerenciada no Unity Catalog
df_bronze_school.write \
    .mode(get_write_mode_by_year(bronze_path_school, year_ref)) \
    .saveAsTable(bronze_path_school)

df_bronze_class.write \
    .mode(get_write_mode_by_year(bronze_path_class, year_ref)) \
    .saveAsTable(bronze_path_class)

# Totais de registros nas tabelas
total_bronze_school = spark.table(bronze_path_school).count()
total_bronze_class = spark.table(bronze_path_class).count()

print(f"\n ✅ Camada BRONZE da Tabela Escola criada!")
print(f"   Registros:  {total_bronze_school:,}")
print(f"   Formato:    Delta Lake (gerenciada pelo Unity Catalog)")
print(f"   Tabela:     {bronze_path_school}")

print(f"\n ✅ Camada BRONZE da Tabela Turma criada!")
print(f"   Registros:  {total_bronze_class:,}")
print(f"   Formato:    Delta Lake (gerenciada pelo Unity Catalog)")
print(f"   Tabela:     {bronze_path_class}")
```

```python
# ================================================
# Segunda Etapa: Definição dos schemas
# ================================================

from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

# Função para aplicar schema a um dataframe utilizando cast()
def apply_schema(df, schema: StructType):
    return df.select([
        f.col(field.name).cast(field.dataType).alias(field.name)
        for field in schema.fields
    ])

schema_school = StructType([
    StructField('NU_ANO_CENSO', IntegerType(), True),
    StructField('NO_ENTIDADE', StringType(), True),
    StructField('CO_ENTIDADE', IntegerType(), True),
    StructField('NO_REGIAO', StringType(), True),
    StructField('CO_REGIAO', IntegerType(), True),
    StructField('NO_UF', StringType(), True),
    StructField('SG_UF', StringType(), True),
    StructField('CO_UF', IntegerType(), True),
    StructField('NO_MUNICIPIO', StringType(), True),
    StructField('CO_MUNICIPIO', IntegerType(), True),
    StructField('NO_DISTRITO', StringType(), True),
    StructField('CO_DISTRITO', IntegerType(), True),
    StructField('NO_REGIAO_ADMINISTRATIVA', StringType(), True),
    StructField('CO_REGIAO_ADMINISTRATIVA', IntegerType(), True),
    StructField('TP_DEPENDENCIA', IntegerType(), True),
    StructField('TP_CATEGORIA_ESCOLA_PRIVADA', IntegerType(), True),
    StructField('TP_LOCALIZACAO', IntegerType(), True),
    StructField('TP_LOCALIZACAO_DIFERENCIADA', IntegerType(), True),
    StructField('DS_ENDERECO', StringType(), True),
    StructField('NU_ENDERECO', StringType(), True),
    StructField('DS_COMPLEMENTO', StringType(), True),
    StructField('NO_BAIRRO', StringType(), True),
    StructField('CO_CEP', StringType(), True),
    StructField('NU_DDD', IntegerType(), True),
    StructField('NU_TELEFONE', IntegerType(), True),
    StructField('LATITUDE', DoubleType(), True),
    StructField('LONGITUDE', DoubleType(), True),
    StructField('TP_SITUACAO_FUNCIONAMENTO', IntegerType(), True),
    StructField('IN_LABORATORIO_CIENCIAS', IntegerType(), True),
    StructField('IN_LABORATORIO_INFORMATICA', IntegerType(), True),
    StructField('IN_QUADRA_ESPORTES', IntegerType(), True),
    StructField('IN_INTERNET', IntegerType(), True),
    StructField('IN_BANDA_LARGA', IntegerType(), True),
    StructField('IN_BIBLIOTECA', IntegerType(), True),
    StructField('IN_REFEITORIO', IntegerType(), True),
    StructField('IN_BANHEIRO', IntegerType(), True),
    StructField('IN_AGUA_POTAVEL', IntegerType(), True),
    StructField('IN_AGUA_INEXISTENTE', IntegerType(), True),
    StructField('IN_ENERGIA_REDE_PUBLICA', IntegerType(), True),
    StructField('IN_ENERGIA_INEXISTENTE', IntegerType(), True),
    StructField('IN_ESGOTO_REDE_PUBLICA', IntegerType(), True),
    StructField('IN_ESGOTO_INEXISTENTE', IntegerType(), True),
])

schema_class = StructType([
    StructField('NU_ANO_CENSO', IntegerType(), True),
    StructField('CO_ENTIDADE', IntegerType(), True),
    StructField('QT_TUR_BAS', IntegerType(), True),
    StructField('QT_TUR_INF', IntegerType(), True),
    StructField('QT_TUR_INF_CRE', IntegerType(), True),
    StructField('QT_TUR_INF_PRE', IntegerType(), True),
    StructField('QT_TUR_FUND', IntegerType(), True),
    StructField('QT_TUR_FUND_AI', IntegerType(), True),
    StructField('QT_TUR_FUND_AF', IntegerType(), True),
    StructField('QT_TUR_MED', IntegerType(), True),
    StructField('QT_TUR_PROF', IntegerType(), True),
    StructField('QT_TUR_EJA', IntegerType(), True),
    StructField('QT_TUR_ESP', IntegerType(), True)
])

df_silver_school = apply_schema(df_silver_school, schema_school)
df_silver_class = apply_schema(df_silver_class, schema_class)

print("=" * 65)
print(f"   SCHEMA - Tabela Escola")
print(f"   Total de colunas: {len(df_silver_school.columns):,}")
print("=" * 65)
df_silver_school.printSchema()

print("=" * 65)
print(f"   SCHEMA - Tabela Turma")
print(f"   Total de colunas: {len(df_silver_class.columns):,}")
print("=" * 65)
df_silver_class.printSchema()
```

## Camada GOLD
Nessa etapa do pipeline, criamos relatórios e métricas de negócio para serem consumidas por áreas analíticas, dashboards e aplicações finais.
A camada Gold representa o nível mais refinado dos dados dentro da arquitetura Lakehouse.

```python
silver_path_school = "workspace.default.silver_escola"
silver_path_class = "workspace.default.silver_turma"

df_silver_school = spark.table(silver_path_school)
df_silver_class = spark.table(silver_path_class)
```

```python
# ============================================================
# GOLD 1: Classificação de infraestrutura e saneamento por estado
# ============================================================

# =========> CÁLCULOS DE PORCENTAGEM

# Aplica expressão para cálculo de porcentagem
def perc(condition):
    return f.round(f.avg(f.when(condition, 1).otherwise(0)) * 100, 2)

# Define as colunas de métricas de infraestrutura e saneamento
metrics = {
    "PERC_POSSUI_AGUA": f.col("IN_AGUA_INEXISTENTE") == 0,
    "PERC_POSSUI_ESGOTO": f.col("IN_ESGOTO_INEXISTENTE") == 0,
    "PERC_POSSUI_BANHEIRO": f.col("IN_BANHEIRO") == 1,
    "PERC_POSSUI_ENERGIA": f.col("IN_ENERGIA_INEXISTENTE") == 0,
    "PERC_POSSUI_LAB_CIENCIAS": f.col("IN_LABORATORIO_CIENCIAS") == 1,
    "PERC_POSSUI_LAB_INFORMATICA": f.col("IN_LABORATORIO_INFORMATICA") == 1,
    "PERC_POSSUI_QUADRA_ESPORTES": f.col("IN_QUADRA_ESPORTES") == 1,
    "PERC_POSSUI_INTERNET": f.col("IN_INTERNET") == 1,
    "PERC_POSSUI_BANDA_LARGA": f.col("IN_BANDA_LARGA") == 1,
    "PERC_POSSUI_BIBLIOTECA": f.col("IN_BIBLIOTECA") == 1,
    "PERC_POSSUI_REFEITORIO": f.col("IN_REFEITORIO") == 1
}

# =========> PESOS PARA SCORE DE CLASSIFICAÇÃO

# Define os pesos por importância (Saneamento > Infraestutura > Lazer)
weights = {
    "PERC_POSSUI_AGUA": 2,
    "PERC_POSSUI_ESGOTO": 2,
    "PERC_POSSUI_ENERGIA": 2,
    "PERC_POSSUI_BANHEIRO": 2,
    "PERC_POSSUI_INTERNET": 1.5,
    "PERC_POSSUI_BANDA_LARGA": 1.5,
    "PERC_POSSUI_LAB_INFORMATICA": 1,
    "PERC_POSSUI_LAB_CIENCIAS": 1,
    "PERC_POSSUI_BIBLIOTECA": 1,
    "PERC_POSSUI_QUADRA_ESPORTES": 0.5,
    "PERC_POSSUI_REFEITORIO": 1
}

# Total dos pesos
total_weight = sum(weights.values())

# =========> DATAFRAME INICIAL: PORCENTAGENS

gold_perc_infra_por_estado = (
    df_silver_school.alias("school")
    .join(df_silver_class.alias("class"), on="CO_ENTIDADE", how="left")
    .filter(
        (f.col("school.NU_ANO_CENSO") == year_ref) &
        (
            f.col('IN_LABORATORIO_CIENCIAS').isNotNull() &
            f.col('IN_LABORATORIO_INFORMATICA').isNotNull() &
            f.col('IN_QUADRA_ESPORTES').isNotNull() &
            f.col('IN_INTERNET').isNotNull() &
            f.col('IN_BANDA_LARGA').isNotNull() &
            f.col('IN_BIBLIOTECA').isNotNull() &
            f.col('IN_REFEITORIO').isNotNull() &
            f.col('IN_BANHEIRO').isNotNull() &
            f.col('IN_AGUA_INEXISTENTE').isNotNull() &
            f.col('IN_ENERGIA_INEXISTENTE').isNotNull() &
            f.col('IN_ESGOTO_INEXISTENTE').isNotNull()
        )
    )
    .groupBy("school.NU_ANO_CENSO", "NO_UF")
    .agg(
        f.count("*").alias("QTD_ESCOLAS"),
        f.sum("QT_TUR_TOTAL").alias("QTD_TURMAS"),
        *[
            perc(condition).alias(col_name)
            for col_name, condition in metrics.items()
        ]
    )
    .orderBy("NO_UF")
)

# =========> DATAFRAME FINAL: SCORE E CLASSIFICAÇÃO

gold_perc_infra_por_estado = (
    gold_perc_infra_por_estado
    .withColumn(
        "SCORE_INFRA",
        f.round(
            sum([f.col(c) * weights[c] for c in weights]) / sum(weights.values()),
            2
        )
    )
    .withColumn(
        "CLASSIFICACAO",
        f.when(f.col("SCORE_INFRA") >= 85, "A")
         .when(f.col("SCORE_INFRA") >= 80, "B")
         .when(f.col("SCORE_INFRA") >= 75, "C")
         .when(f.col("SCORE_INFRA") >= 70, "D")
         .otherwise("E")
    )
)

# =======> SALVANDO A TABELA

# Define o modo de escrita
def get_write_mode_by_year(target_path, year_ref):
    year_on_table = spark.table(target_path).select("NU_ANO_CENSO").filter(f.col("NU_ANO_CENSO") == year_ref).limit(1)
    if year_on_table.count() == 1:
        return "overwrite"
    else:
        return "append"

# Caminho para a tabela
path = "workspace.default.gold_perc_infra_por_estado"

# Salva a tabela no Unity Catalog
gold_perc_infra_por_estado.write.mode(get_write_mode_by_year(path, year_ref)).saveAsTable(path)

# =======> VISUALIZAÇÃO

display(gold_perc_infra_por_estado)
```

```python
# ============================================================
# GOLD 2: Quantidade de escolas sem dados de infraestrutura e saneamento
# ============================================================

gold_infra_sem_dados = df_silver_school \
        .filter(
            (f.col('NU_ANO_CENSO') == year_ref) & (
            f.col('IN_LABORATORIO_CIENCIAS').isNull() |
            f.col('IN_LABORATORIO_INFORMATICA').isNull() |
            f.col('IN_QUADRA_ESPORTES').isNull() |
            f.col('IN_INTERNET').isNull() |
            f.col('IN_BANDA_LARGA').isNull() |
            f.col('IN_BIBLIOTECA').isNull() |
            f.col('IN_REFEITORIO').isNull() |
            f.col('IN_BANHEIRO').isNull() |
            f.col('IN_AGUA_POTAVEL').isNull() |
            f.col('IN_AGUA_INEXISTENTE').isNull() |
            f.col('IN_ENERGIA_REDE_PUBLICA').isNull() |
            f.col('IN_ENERGIA_INEXISTENTE').isNull() |
            f.col('IN_ESGOTO_REDE_PUBLICA').isNull() |
            f.col('IN_ESGOTO_INEXISTENTE').isNull() 
            )
            ) \
        .groupBy("NU_ANO_CENSO", "NO_UF") \
        .agg(
            f.count("*").alias("QTD_SEM_DADOS_INFRA")
            ) \
        .orderBy("NO_UF")

# =======> SALVANDO A TABELA

# Define o modo de escrita
def get_write_mode_by_year(target_path, year_ref):
    year_on_table = spark.table(target_path).select("NU_ANO_CENSO").filter(f.col("NU_ANO_CENSO") == year_ref).limit(1)
    if year_on_table.count() == 1:
        return "overwrite"
    else:
        return "append"

# Caminho para a tabela
path = "workspace.default.gold_infra_sem_dados"

# Salva a tabela no Unity Catalog
gold_infra_sem_dados.write.mode(get_write_mode_by_year(path, year_ref)).saveAsTable(path)

# =======> VISUALIZAÇÃO

display(gold_infra_sem_dados)
```
