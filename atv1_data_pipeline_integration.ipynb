## Iniciando projeto

```python
df_curso_tecnico = spark.read.csv("/Volumes/workspace/default/censo_escolar/Tabela_Curso_Tecnico_2025.csv", header=True, inferSchema=True, sep=';')
df_escola = spark.read.csv("/Volumes/workspace/default/censo_escolar/Tabela_Escola_2025.csv", header=True, inferSchema=True, sep=';')
df_matricula = spark.read.csv("/Volumes/workspace/default/censo_escolar/Tabela_Matricula_2025.csv", header=True, inferSchema=True, sep=';')
df_turma = spark.read.csv("/Volumes/workspace/default/censo_escolar/Tabela_Turma_2025.csv", header=True, inferSchema=True, sep=';')
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

schema = StructType([
    StructField('NU_ANO_CENSO', IntegerType(), True),
    StructField('NO_REGIAO', StringType(), True),
    StructField('NO_UF', StringType(), True),


```

```python
# Distribuição de escolas por UF
uf_dist = df_escola.groupBy('NO_UF').count().orderBy('count', ascending=False)
display(uf_dist)

# Proporção de escolas por dependência administrativa
dep_dist = df_escola.groupBy('TP_DEPENDENCIA').count().orderBy('count', ascending=False)
display(dep_dist)

# Quantidade de escolas urbanas vs rurais
loc_dist = df_escola.groupBy('TP_LOCALIZACAO').count().orderBy('count', ascending=False)
display(loc_dist)
```

```python
# Totais de matrículas por etapa
mat_bas = df_matricula.agg({'QT_MAT_BAS': 'sum'}).collect()[0][0]
mat_inf = df_matricula.agg({'QT_MAT_INF': 'sum'}).collect()[0][0]
mat_fund = df_matricula.agg({'QT_MAT_FUND': 'sum'}).collect()[0][0]
mat_med = df_matricula.agg({'QT_MAT_MED': 'sum'}).collect()[0][0]
mat_prof = df_matricula.agg({'QT_MAT_PROF': 'sum'}).collect()[0][0]
mat_eja = df_matricula.agg({'QT_MAT_EJA': 'sum'}).collect()[0][0]
display(spark.createDataFrame(
    [(mat_bas, mat_inf, mat_fund, mat_med, mat_prof, mat_eja)],
    ['tot_basica', 'tot_infantil', 'tot_fundamental', 'tot_medio', 'tot_profissional', 'tot_eja']
))

# Perfis de matrícula por sexo
mat_sexo = df_matricula.select('QT_MAT_BAS_FEM', 'QT_MAT_BAS_MASC').agg(
    {'QT_MAT_BAS_FEM': 'sum', 'QT_MAT_BAS_MASC': 'sum'})
display(mat_sexo)
```

```python
# Quantidade de turmas por escola
turma_escola = df_turma.groupBy('CO_ENTIDADE').agg({'QT_TUR_BAS': 'sum'}).orderBy('sum(QT_TUR_BAS)', ascending=False)
display(turma_escola)

# Relação alunos/turma (cruzamento sugerido)
rel_aluno_turma = df_matricula.join(df_turma, 'CO_ENTIDADE').select('CO_ENTIDADE', 'QT_MAT_BAS', 'QT_TUR_BAS').withColumn('alunos_por_turma', df_matricula.QT_MAT_BAS / df_turma.QT_TUR_BAS)
display(rel_aluno_turma)
```

```python
# Matrículas em cursos técnicos por área
mat_area = df_curso_tecnico.groupBy('NO_AREA_CURSO_PROFISSIONAL').agg({'QT_MAT_CURSO_TEC': 'sum'}).orderBy('sum(QT_MAT_CURSO_TEC)', ascending=False)
display(mat_area)

# Oferta de cursos técnicos por UF (se o campo NO_UF existir)
cursos_uf = df_curso_tecnico.groupBy('NO_UF').agg({'QT_CURSO_TEC': 'sum'}).orderBy('sum(QT_CURSO_TEC)', ascending=False)
display(cursos_uf)
```
