## Extraindo os Datasets
Explicação...

```python
# Obtém o ano atual -1 para referência na coleta dos dados
year_ref = datetime.now().year - 1

# Caminho do catálogo onde os arquivos CSV serão armazenados
volume_path = "/Volumes/workspace/default/censo_escolar/"

# Lista das tabelas a serem carregadas no ambiente
target_tables = [
    f"Tabela_Turma_{year_ref}.csv",
    f"Tabela_Escola_{year_ref}.csv"
]
```

```python
%sql
-- Cria o volume de destino dos arquivos CSV caso ainda não exista
CREATE VOLUME IF NOT EXISTS workspace.default.censo_escolar
```

```python
volume_files = os.listdir(volume_path) # Obtém os arquivos que já estão no catálogo

result = all(item in volume_files for item in target_tables) # Verifica se as tabelas alvo já estão presentes no catálogo
print(result)
```

Upload e carregamento do arquivo ZIP do governo (Caso os dados não estejam carregados no Catálogo)

```python
if not result:

    data_url = f"https://download.inep.gov.br/dados_abertos/microdados_censo_escolar_{year_ref}_.zip" # URL de dowload no site do governo
    zip_path = f"/tmp/microdados_censo_escolar_{year_ref}_.zip" # Caminho de destino para o zip à ser baixado

    with requests.get(data_url, stream=True) as r:
        r.raise_for_status()
        with open(zip_path, "wb") as f:
            for chunk in r.iter_content(8192):
                f.write(chunk)

    print("Arquivo ZIP baixado")
```

```python
if not result:

    extract_path = "/tmp/extracted" # Caminho de destino para os dados extraídos do zip

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

    print("Arquivos extraídos:", os.listdir(extract_path))
```

```python
if not result:

    tables_path = extract_path + f"/microdados_censo_escolar_{year_ref}/dados/" # Caminho onde estão os arquivos CSV

    # Lista os arquivos CSV encontrados
    for file in os.listdir(tables_path):
        if file.endswith(".csv"):
            csv_file = os.path.join(tables_path, file)
            print("CSV encontrado:", csv_file)
```

```python
if not result:
      
      import shutil

      for table in target_tables:
            source_path = tables_path + table # Caminho de origem
            destination_path = volume_path + table # Caminho de destino
            
            # Remove se já existir
            if os.path.exists(destination_path):
                  os.remove(destination_path)

            # Copia o CSV do /temp para o catálogo
            shutil.copy(source_path, destination_path)

            print(
            f"""
      Arquivo CSV adicionado ao catálogo com sucesso!!
      Arquivo: {source_path}
      Caminho: {destination_path}
            """
            )
```

```python
# Remove os arquivos da pasta temporária
try:
    os.remove(zip_path)
    shutil.rmtree(extract_path)
    print("Arquivos temporários removidos com sucesso!")
except:
    print("Os arquivos já não existem na pasta temporária")
```

### Camada BRONZE
Salvando dados brutos com metadados de controle

```python
# Adiciona 3 colunas de controle
df_bronze_school = df_raw_school \
    .withColumn("_ingestao_timestamp", f.current_timestamp()) \
    .withColumn("_fonte", f.lit(f"gov/Tabela_Escola{year_ref}.csv")) \
    .withColumn("_pipeline_version", f.lit("1.0.0"))

df_bronze_class = df_raw_class \
    .withColumn("_ingestao_timestamp", f.current_timestamp()) \
    .withColumn("_fonte", f.lit(f"gov/Tabela_Turma{year_ref}.csv")) \
    .withColumn("_pipeline_version", f.lit("1.0.0"))
```

```python
# ================================================
# Segunda Etapa: Definição dos schemas
# ================================================

from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

# Função para aplicar schema a um dataframe utilizando cast()
def apply_schema(df, schema: StructType):
    return df.select([
        f.col(field.name).cast(field.dataType).alias(field.name)
        for field in schema.fields
    ])

schema_school = StructType([
    StructField('NU_ANO_CENSO', IntegerType(), True),
    StructField('NO_ENTIDADE', StringType(), True),
    StructField('CO_ENTIDADE', IntegerType(), True),
    StructField('NO_REGIAO', StringType(), True),
    StructField('CO_REGIAO', IntegerType(), True),
    StructField('NO_UF', StringType(), True),
    StructField('SG_UF', StringType(), True),
    StructField('CO_UF', IntegerType(), True),
    StructField('NO_MUNICIPIO', StringType(), True),
    StructField('CO_MUNICIPIO', IntegerType(), True),
    StructField('NO_DISTRITO', StringType(), True),
    StructField('CO_DISTRITO', IntegerType(), True),
    StructField('NO_REGIAO_ADMINISTRATIVA', StringType(), True),
    StructField('CO_REGIAO_ADMINISTRATIVA', IntegerType(), True),
    StructField('TP_DEPENDENCIA', IntegerType(), True),
    StructField('TP_CATEGORIA_ESCOLA_PRIVADA', IntegerType(), True),
    StructField('TP_LOCALIZACAO', IntegerType(), True),
    StructField('TP_LOCALIZACAO_DIFERENCIADA', IntegerType(), True),
    StructField('DS_ENDERECO', StringType(), True),
    StructField('NU_ENDERECO', StringType(), True),
    StructField('DS_COMPLEMENTO', StringType(), True),
    StructField('NO_BAIRRO', StringType(), True),
    StructField('CO_CEP', StringType(), True),
    StructField('NU_DDD', IntegerType(), True),
    StructField('NU_TELEFONE', IntegerType(), True),
    StructField('LATITUDE', DoubleType(), True),
    StructField('LONGITUDE', DoubleType(), True),
    StructField('TP_SITUACAO_FUNCIONAMENTO', IntegerType(), True),
    StructField('IN_LABORATORIO_CIENCIAS', IntegerType(), True),
    StructField('IN_LABORATORIO_INFORMATICA', IntegerType(), True),
    StructField('IN_QUADRA_ESPORTES', IntegerType(), True),
    StructField('IN_INTERNET', IntegerType(), True),
    StructField('IN_BANDA_LARGA', IntegerType(), True),
    StructField('IN_BIBLIOTECA', IntegerType(), True),
    StructField('IN_REFEITORIO', IntegerType(), True),
    StructField('IN_BANHEIRO', IntegerType(), True),
    StructField('IN_AGUA_POTAVEL', IntegerType(), True),
    StructField('IN_AGUA_INEXISTENTE', IntegerType(), True),
    StructField('IN_ENERGIA_REDE_PUBLICA', IntegerType(), True),
    StructField('IN_ENERGIA_INEXISTENTE', IntegerType(), True),
    StructField('IN_ESGOTO_REDE_PUBLICA', IntegerType(), True),
    StructField('IN_ESGOTO_INEXISTENTE', IntegerType(), True),
])

schema_class = StructType([
    StructField('NU_ANO_CENSO', IntegerType(), True),
    StructField('CO_ENTIDADE', IntegerType(), True),
    StructField('QT_TUR_BAS', IntegerType(), True),
    StructField('QT_TUR_INF', IntegerType(), True),
    StructField('QT_TUR_INF_CRE', IntegerType(), True),
    StructField('QT_TUR_INF_PRE', IntegerType(), True),
    StructField('QT_TUR_FUND', IntegerType(), True),
    StructField('QT_TUR_FUND_AI', IntegerType(), True),
    StructField('QT_TUR_FUND_AF', IntegerType(), True),
    StructField('QT_TUR_MED', IntegerType(), True),
    StructField('QT_TUR_PROF', IntegerType(), True),
    StructField('QT_TUR_EJA', IntegerType(), True),
    StructField('QT_TUR_ESP', IntegerType(), True)
])

# schema_matricula = StructType([
#     StructField('NU_ANO_CENSO', IntegerType(), True),
#     StructField('CO_ENTIDADE', IntegerType(), True),
#     StructField('QT_MAT_BAS', IntegerType(), True),
#     StructField('QT_MAT_INF', IntegerType(), True),
#     StructField('QT_MAT_INF_CRE', IntegerType(), True),
#     StructField('QT_MAT_INF_PRE', IntegerType(), True),
#     StructField('QT_MAT_FUND', IntegerType(), True),
#     StructField('QT_MAT_FUND_AI', IntegerType(), True),
#     StructField('QT_MAT_FUND_AF', IntegerType(), True),
#     StructField('QT_MAT_MED', IntegerType(), True),
#     StructField('QT_MAT_PROF', IntegerType(), True),
#     StructField('QT_MAT_EJA', IntegerType(), True),
#     StructField('QT_MAT_ESP', IntegerType(), True)
# ])

df_silver_school = apply_schema(df_silver_school, schema_school)
df_silver_class = apply_schema(df_silver_class, schema_class)

print("=" * 65)
print(f"   SCHEMA - Tabela Escola")
print(f"   Total de colunas: {len(df_silver_school.columns):,}")
print("=" * 65)
df_silver_school.printSchema()

print("=" * 65)
print(f"   SCHEMA - Tabela Turma")
print(f"   Total de colunas: {len(df_silver_class.columns):,}")
print("=" * 65)
df_silver_class.printSchema()
```

```python
# ============================================================
# GOLD 1: Porcentagem de infraestrutura e saneamento por estado
# ============================================================

gold_perc_infra_por_estado = df_silver_school \
        .filter(
            (f.col('NU_ANO_CENSO') == year_ref) & (
            f.col('IN_LABORATORIO_CIENCIAS').isNotNull() &
            f.col('IN_LABORATORIO_INFORMATICA').isNotNull() &
            f.col('IN_QUADRA_ESPORTES').isNotNull() &
            f.col('IN_INTERNET').isNotNull() &
            f.col('IN_BANDA_LARGA').isNotNull() &
            f.col('IN_BIBLIOTECA').isNotNull() &
            f.col('IN_REFEITORIO').isNotNull() &
            f.col('IN_BANHEIRO').isNotNull() &
            f.col('IN_AGUA_POTAVEL').isNotNull() &
            f.col('IN_AGUA_INEXISTENTE').isNotNull() &
            f.col('IN_ENERGIA_REDE_PUBLICA').isNotNull() &
            f.col('IN_ENERGIA_INEXISTENTE').isNotNull() &
            f.col('IN_ESGOTO_REDE_PUBLICA').isNotNull() &
            f.col('IN_ESGOTO_INEXISTENTE').isNotNull()
            ) 
            ) \
        .groupBy("NU_ANO_CENSO", "NO_UF") \
        .agg(
            f.count("*").alias("QTD_ESCOLAS"),
            f.round(((f.count(f.col("IN_AGUA_INEXISTENTE") == 0) / f.count("*")) * 100), 2).alias("PERC_POSSUI_AGUA"),
            f.round(((f.count(f.col("IN_ESGOTO_INEXISTENTE") == 0) / f.count("*")) * 100), 2).alias("PERC_POSSUI_ESGOTO"),
            f.round(((f.count(f.col("IN_BANHEIRO") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_BANHEIRO"),
            f.round(((f.count(f.col("IN_ENERGIA_INEXISTENTE") == 0) / f.count("*")) * 100), 2).alias("PERC_POSSUI_ENERGIA"),
            f.round(((f.count(f.col("IN_LABORATORIO_CIENCIAS") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_LAB_CIENCIAS"),
            f.round(((f.count(f.col("IN_LABORATORIO_INFORMATICA") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_LAB_INFORMATICA"),
            f.round(((f.count(f.col("IN_QUADRA_ESPORTES") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_QUADRA_ESPORTES"),
            f.round(((f.count(f.col("IN_INTERNET") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_INTERNET"),
            f.round(((f.count(f.col("IN_BANDA_LARGA") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_BANDA_LARGA"),
            f.round(((f.count(f.col("IN_BIBLIOTECA") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_BIBLIOTECA"),
            f.round(((f.count(f.col("IN_REFEITORIO") == 1) / f.count("*")) * 100), 2).alias("PERC_POSSUI_REFEITORIO")
            ) \
        .orderBy("NO_UF")

gold_perc_infra_por_estado.write.mode("overwrite").saveAsTable("workspace.default.gold_perc_infra_por_estado")

display(gold_perc_infra_por_estado)
```
